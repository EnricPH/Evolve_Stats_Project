**`Valores Nulos`**: En caso de que el conjunto de datos contuviese valores nulos, se deber√≠a estudiar si son de tipo:
- Missing at Random (*MAR*): Hay relaci√≥n entre la variable nula respecto a otra/s variable/s.

- Missing Completely at Random (*MCAR*): No hay relaci√≥n entre la variable nula respecto a otras variables
    - Eg: Sensor de temperatura falla por motivos de isntalaci√≥n, causando valores nulos en valores de temperatura aleatorios.
        
- Missing Not at Random (*MNAR*): Las variables nulas est√°n relacionadas con el valor de la propia variable en s√≠.
    - Eg: Sensor de temperatura falla cuando la temperatura es demasiado alta.

Dependiendo del tipo de valor nulo, de la import√°ncia de la variable y de la cantidad de valores nulos, se pueden aplicar distintas t√©cnicas de imputaci√≥n o simplemente la eliminaci√≥n de las instancias o de la variable completa.



#### 1.2.  **Clasificaci√≥n de Variables:**
En la secci√≥n anterior se ha visto que todas las variables eran de tipo *int*, ahora bien ¬øcu√°les son discretas, cu√°les son binarias y cu√°les podrian pertenecer a variables categoricas ordinales?

Para las variables numericas ordinales, se aprovechar√° el hecho de que ya se sabe, gracias a la descripci√≥n previa del *dataset*, que todas las columnas num√©ricas ordinales no contienen m√°s de 5 "categorias".

````python
def variables_por_tipo(df, tipo):
    """
    Filtra columnas num√©ricas de un DataFrame seg√∫n su tipo:
    'binario', 'ordinal' o 'discreto'.

    Par√°metros
    ----------
    df : pd.DataFrame
        DataFrame de entrada.
    tipo : str
        Tipo de variable a seleccionar. Debe ser uno de:
        {'binario', 'ordinal', 'discreto'}.

    Returns
    -------
    pd.DataFrame
        DataFrame con solo las columnas del tipo solicitado.

    Raises
    ------
    TypeError
        Si df no es un DataFrame de pandas.
    ValueError
        Si tipo no es uno de los valores permitidos.
    """

    # Validaci√≥n de errores
    if not isinstance(df, pd.DataFrame):
        raise TypeError("El primer argumento 'df' debe ser un pandas DataFrame.")

    tipos_validos = {'binario', 'ordinal', 'discreto'}
    if tipo not in tipos_validos:
        raise ValueError(
            f"Tipo '{tipo}' no v√°lido. Debe ser uno de {tipos_validos}."
        )

    # Selecci√≥n solo de columnas num√©ricas
    columnas_seleccionadas = []

    for col in df.columns:
        valores_unicos = len(df[col].unique())

        # BINARIAS
        if tipo == 'binario':
            if valores_unicos == 2:
                columnas_seleccionadas.append(col)

        # ORDINALES (discretas con pocas categor√≠as)
        elif tipo == 'ordinal':
            if valores_unicos > 2 and valores_unicos <= 5:
                columnas_seleccionadas.append(col)

        # DISCRETAS (mayor cardinalidad)
        elif tipo == 'discreto':
            if valores_unicos > 5:
                columnas_seleccionadas.append(col)

    # Output
    print(f"\nVariables de tipo {tipo} seleccionadas:")
    for c in columnas_seleccionadas:
        print('\t¬∑', c)

    return df[columnas_seleccionadas]

# Construios el dataframes para cada uno de los tipos de variables
bin_df = variables_por_tipo(df, 'binario')
display(bin_df.head())

ord_df = variables_por_tipo(df, 'ordinal')
display(ord_df.head())

disc_df = variables_por_tipo(df, 'discreto')
display(disc_df.head())

````

Separar las variables n√∫mericas de esta forma (15 binarias, 3 ordinales, 12 discretas), facilitar√° el estudio para pasos posteriores, ya que tratar de analizar las 30 variables a la vez ser√≠a contraproducente.

#### 1.3.  **Estad√≠sticos Descriptivos:**

````python
X = disc_df[['crp_level']]
y = disc_df[['fev1_x10']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Scikit-Learn
lr_model = LinearRegression()
lr_model.fit(X_train,y_train)

# Par√°metros aprendidos
a0 = lr_model.intercept_[0]
a1 = lr_model.coef_[0][0]

# Predicciones
y_pred = lr_model.predict(X_test)

# M√©tricas
rmse = mean_squared_error(y_test, y_pred)**(1/2)
r2 = r2_score(y_test, y_pred)

print('Intercept', a0)
print('Slope', a1)
print('RMSE', rmse)
print('R_Squared', r2)

plt.figure(figsize=(12,6))
plt.scatter(X_train, y_train, label='Datos de Entrenamiento', alpha=0.7)
plt.scatter(X_test, y_test, label='Datos de Test', color='orange', alpha=0.3)
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regresi√≥n')
plt.legend()
plt.xlabel('crp level (mg/L)')
plt.ylabel('fev1_x10 (litros/segundo)*10')
plt.show()
````

**COMPROBACI√ìN DE LAS SUPOSICIONES**

Las suposiciones generalmente se comprueban usando el conjunto de entrenamiento 

- **Linealidad**
- **Normalidad en los Errores**
- **Independecia de los Errores**
- **Varianza Constante en los Errores**

````python
import scipy.stats as stats

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Error o residuos
y_pred_train = lr_model.predict(X_train)
residuals = y_train.values - y_pred_train

#Paso necesario de conversion a pd.Series
y_pred_series = pd.Series(y_pred_train.ravel(), name='y_pred')
residuals = pd.Series(residuals.ravel(), name='residuals')

# 1. Linearity: Residuals vs Fitted
axes[0,0].scatter(y_pred_series, residuals, alpha=0.6)
axes[0,0].axhline(y=0, color='red', linestyle='--')
axes[0,0].set_xlabel('Fitted Values')
axes[0,0].set_ylabel('Residuals')
axes[0,0].set_title('Residuals vs Fitted\n(Checking Linearity & Homoscedasticity)')

# 2. Normality: Q-Q Plot
stats.probplot(residuals, dist="norm", plot=axes[0,1])
axes[0,1].set_title('Q-Q Plot\n(Checking Normality of Residuals)')

# 3. Independence: Residuals vs Order
axes[1,0].plot(residuals, 'o-', alpha=0.6)
axes[1,0].axhline(y=0, color='red', linestyle='--')
axes[1,0].set_xlabel('Observation Order')
axes[1,0].set_ylabel('Residuals')
axes[1,0].set_title('Residuals vs Order\n(Checking Independence)')

# 4. Homoscedasticity: Scale-Location Plot
standardized_residuals = np.abs(residuals / np.std(residuals))
axes[1,1].scatter(y_pred_series, standardized_residuals, alpha=0.6)
axes[1,1].axhline(y=0, color='red', linestyle='--')
axes[1,1].set_xlabel('Fitted Values')
axes[1,1].set_ylabel('‚àö|Standardized Residuals|')
axes[1,1].set_title('Scale-Location Plot\n(Checking Homoscedasticity)')

plt.tight_layout()
plt.show()

## STATISTICAL TESTS FOR ASSUMPTIONS
print("\n" + "="*50)
print("STATISTICAL TESTS FOR ASSUMPTIONS")
print("="*50)

# Normality test (Shapiro-Wilk)
shapiro_stat, shapiro_p = stats.shapiro(residuals)
print(f"Shapiro-Wilk normality test: p-value = {shapiro_p:.4f}")
print("‚Üí Residuals are normal" if shapiro_p > 0.05 else "‚Üí Residuals are NOT normal")

# Homoscedasticity test (Breusch-Pagan would be better, but simpler approach)
# Check correlation between fitted values and squared residuals
corr_coef, corr_p = stats.pearsonr(y_pred_series, residuals**2)
print(f"Correlation between fitted and squared residuals: {corr_coef:.4f} (p={corr_p:.4f})")
print("‚Üí Homoscedasticity likely holds" if corr_p > 0.05 else "‚Üí Possible heteroscedasticity")

# Independence (Durbin-Watson)
from statsmodels.stats.stattools import durbin_watson
dw_stat = durbin_watson(residuals)
print(f"Durbin-Watson statistic: {dw_stat:.4f}")
print("‚Üí No significant autocorrelation" if 1.5 < dw_stat < 2.5 else "‚Üí Possible autocorrelation")

print("\n" + "="*50)
print("CONCLUSION")
print("="*50)
print("‚úì Manual OLS and scikit-learn produce IDENTICAL results (up to numerical precision)")
print("‚úì All OLS assumptions are satisfied in our generated data")
print("‚úì The model successfully recovers the true underlying relationship")
````

#### 2.4. **Regresi√≥n logistica:**

A pesar de su nombre, la **Regresi√≥n Logistica** es un modelo de **clasificaci√≥n**, no de **regresi√≥n**, por lo que en este caso si podemos predecir sobre la variable objetivo `lung_cancer_risk`.

Donde la probabilidad de que una instancia de positivo en la variable objetivo, viene dada por:

$$y = \frac{e^{\beta_0 + \beta_1 X}}{1 + e^{\beta_0 + \beta_1 X}}$$

Siendo $\beta_0$ el intercepto y $\beta_1$ la pendiente.

````python
# Variables
X_log = disc_df[['fev1_x10']]
y_log = bin_df['lung_cancer_risk']  # Series, no DataFrame

# Train / Test
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(
    X_log, y_log, test_size=0.3, random_state=1
)

# Modelo
logr_model = LogisticRegression()
logr_model.fit(X_train_log, y_train_log)

# Predicciones
y_pred_log = logr_model.predict(X_test_log)
y_prob_log = logr_model.predict_proba(X_test_log)[:, 1]


# Par√°metros
b0 = logr_model.intercept_[0]
b1 = logr_model.coef_[0][0]

# M√©tricas de regresi√≥n
mse_log = mean_squared_error(y_test_log, y_pred_log)
r2_log = r2_score(y_test_log, y_pred_log)

print('Intercept', b0)
print('Slope', b1)

# Ordenar para dibujar la curva correctamente
order = np.argsort(X_test_log.values.flatten())
X_sorted = X_test_log.values.flatten()[order]
y_prob_sorted = y_prob_log[order]

# Plot
plt.figure(figsize=(12,6))

plt.scatter(X_log, y_log, label='Datos')
plt.plot(X_sorted, y_prob_sorted, color='red', linewidth=2, label='Probabilidad (Log√≠stica)', alpha=0.5)
plt.axvline(x=29, color = 'red')
plt.xlabel('fev1_x10 (litros/segundo) √ó 10')
plt.ylabel('Probabilidad de Lung Cancer Risk')
plt.legend()
plt.title('Regresi√≥n Log√≠stica con fev1_x10')
plt.show()
````

Para los problemas de **clasificaci√≥n** existen otras m√©tricas de evaluaci√≥n de modelos, la mayor√≠a basadas en 4 valores (Teniendo en cuenta que el objetivo es predecir con exito `lung_cancer_risk = 1`):
- **Positivos Reales (TP)**: Individuos de la Clase 1 predecidos como Clase 1
- **Negativos Reales (TN)**: Individuos de la Clase 0 predecidos como Clase 0
- **Positivos Falsos (FP)**: Individuos de la Clase 0 predecidos como Clase 1
- **Neagtivos Falsos (FN)**: Individuos de la Clase 1 predecidos como Clase 0

Estos 4 valores forman la llamada **Matriz de Confusi√≥n**

|   | Real 0 | Real 1 |
|---|---|---|
| **Predicci√≥n 0** | TN  | FN  |
| **Predicci√≥n 1** | FP  | TP  |


Las metricas de evlauaci√≥n m√°s comunes son:

- $Accuracy = \frac{TP+TN}{TP+FP+TN+FN}$: Mide la capacidad de acierto general del modelo

- $Precision = \frac{FP}{TP+FP}$: Mide la capacidad de acierto de la clase objetivo. De todos los clasificacdos como Clase 1, ¬øCu√°ntos realmente lo son?

- $Recall = \frac{FN}{TP+FN}$: Mide la capacidad de detecci√≥n de la clase objetivo. De todos los individuos de Clase 1, ¬øCu√°ntos se puedieron detectar?

- $F1\;Score = \frac{2 Precision Recall}{Precision+Recall}$: Mide el balance entre detecci√≥n (*Recall*) y acierto (*Precision*)

Hay otras m√©tricas como las ***specificity*** o ***ROC-AUC***, pero con las anteriores es m√°s que suficiente en este caso.


Para saber que metricas usar en este caso se debe tener en cuenta que la variable objetivo est√° **desbalanceada** (75% Clase 0 vs 25% Clase 1), por lo que usar la *Accuracy* como referencia no es una buena idea, ya que esta mide todos los aciertos indistintivamente de la clase, por lo que un valor alto de *Accuracy* puede infromar que el modelo capta bien los individuos sanos (la mayor√≠a) pero no los otros.\
En este contexto se necesita un balance entre *Precision* y *Recall*, ja que queremos el menor n√∫mero de **falsos positivos** as√≠ como el menor n√∫mero de **falsos negativos** possible.



````python
precision_fev1 = precision_score(y_test_log, y_pred_log)
recall_fev1 = recall_score(y_test_log, y_pred_log)
f1_score_fev1 = f1_score(y_test_log, y_pred_log)

print(f'Precision: {precision_fev1:.4f}')
print(f'Recall: {recall_fev1:.4f}')
print(f'F1 Score: {f1_score_fev1:.4f}')
````

Con tan solo una variable, `fev1_x10`, el modelo es capaz de detectar un 78,80% de los individuos de clase 1, `lung_cancer_risk = 1`, y de todos los individuos predecidos como propensos a riesgo un 90% realmente lo son mientras que el 10% restante son **Falsos positivos**.

## PARTE 3: Regresi√≥n Lineal "From Scratch" (Datos Simulados)

**Objetivo:** Demostrar que entiendes la matem√°tica detr√°s de `LinearRegression()` implement√°ndola con NumPy.

Amplia con:

- **Predicci√≥n y evaluaci√≥n**
  Utiliza tu implementaci√≥n para:
  - Realizar predicciones
  - Calcular manualmente el **MSE** y el **R¬≤**, sin usar funciones de Scikit-Learn.

- **Comparaci√≥n con Scikit-Learn**
  Entrena un modelo `LinearRegression()` sobre los mismos datos y compara:
  - Coeficientes
  - M√©tricas de evaluaci√≥n  
  Discute posibles diferencias num√©ricas.


````python
import numpy as np
import matplotlib.pyplot as plt

# --- NO TOCAR: GENERACI√ìN DE DATOS ---
np.random.seed(42)
X_mock = 2 * np.random.rand(100, 1)
y_mock = 4 + 3 * X_mock + np.random.randn(100, 1)
# -----------------------------------

def evolve_linear_regression(X, y, iteraciones=16, eta=0.1):
    """
    Calcula la pendiente (w), el intercepto (beta_0),
    el error MSE (mse) y R^2 (r_squared)
    usando las f√≥rmulas de OLS con NumPy.
    Returns: (b, w, b_history, w_history, mse_history, r_squared_history)
    """

    # TU C√ìDIGO AQU√ç
    #Generacion aleatoria de parametros (numpy)
    w = np.random.randn()
    b = np.random.randn()

    n = len(X)

    #Arrays vacios para almacenar los par√°metros y error en cada iteraci√≥n
    w_history = []
    b_history = []
    mse_history = []
    r2_history = []

    #Total Sum of Squares (TSS) constante
    tss = np.sum((y - np.mean(y)) ** 2)

    for i in range(iteraciones):
        #Calcular valor predecido y error
        y_pred = b + w * X
        error = y - y_pred

        #Calculo derivadas de descenso de gradiente
        dw = (-2 / n) * np.sum(X * error)
        db = (-2 / n) * np.sum(error)

        #Ajustar parametros
        w = w - eta * dw
        b = b - eta * db

        # M√©tricas
        mse = np.mean(error ** 2)
        rss = np.sum(error ** 2)
        r2 = 1 - rss / tss

        #Guardar c√°lculos
        w_history.append(w)
        b_history.append(b)
        mse_history.append(mse)
        r2_history.append(r2)

    return b, w, b_history, w_history, mse_history, r2_history


# Ejecutamos el modelo
b, w, b_hist, w_hist, mse_history, r2_history = evolve_linear_regression(X_mock, y_mock)


# Scikit-Learn
lr_model = LinearRegression()
lr_model.fit(X_mock, y_mock)

# Par√°metros aprendidos
b_sklearn = lr_model.intercept_[0]
w_sklearn = lr_model.coef_[0][0]

# Predicciones
y_pred_sklearn = lr_model.predict(X_mock)

# M√©tricas
mse_sklearn = mean_squared_error(y_mock, y_pred_sklearn)
r2_sklearn = r2_score(y_mock, y_pred_sklearn)


regresiones = pd.DataFrame({'Modelos':['Tu Modelo', 'Modelo Scikit-Learn'], 
        'Intercepto (b)':[b, b_sklearn], 
        'Pendiente estimada':[w, w_sklearn],
        'Error (MSE)': [mse_history[-1], mse_sklearn],
        'Varianza expicada (R_Squared)':[r2_history[-1], r2_sklearn]})

display(regresiones)

#Visualizaci√≥n
plt.figure(figsize=(7, 5))
plt.scatter(X_mock, y_mock, color='blue', label='Datos')

# Rectas intermedias
c = 0
for bi, wi in zip(b_hist[:-1], w_hist[:-1]):
    if c==0:
        plt.plot(X_mock, bi + wi * X_mock, color='gray', alpha=0.3, label='Iteraciones de tu Regresi√≥n')
        c=1
    else:
        plt.plot(X_mock, bi + wi * X_mock, color='gray', alpha=0.3)

# Recta final manual vs LinearRegression() de Scikit-Learn
plt.plot(X_mock, b + w * X_mock, color='red', linewidth=2, label='Tu Regresi√≥n')
plt.plot(X_mock, y_pred_sklearn, color='green', linewidth=2, label='Regresi√≥n scikit-learn')
plt.title('Evoluci√≥n de la Regresi√≥n Lineal')
plt.legend()

plt.tight_layout()
plt.show()

````

Los resultados obtenidos por ambos modelos son muy similares, tanto en los par√°metros estimados como en las m√©tricas de ajuste:

Las peque√±as discrepancias entre valores son esperables y se explican por el **m√©todo de optimizaci√≥n empleado** en cada enfoque.

**``Modelo Manual`` (Descenso de Gradiente)**

- Se basa en un **proceso iterativo**.
- La soluci√≥n final depende de:
  - el n√∫mero de iteraciones,
  - la tasa de aprendizaje, *learning rate* (Œ∑),
  - los valores iniciales aleatorios de los par√°metros.
- Si el algoritmo no alcanza una convergencia completa, los coeficientes obtenidos son una **aproximaci√≥n** al m√≠nimo del error.

Por otro lado el modelo de **`LinearRegression` de Scikit-Learn**

- Utiliza una **soluci√≥n cerrada (Ordinary Least Squares)**, no iterativa.
- Calcula directamente el **m√≠nimo global exacto** del error cuadr√°tico.

Como consecuencia, los resultados son **m√°s estables, deterministas y reproducibles**.


En general la soluci√≥n que implementa el modelo de `LinearRegression` es m√°s segura, ya que no depende de par√°metros como el *learning rate* el cual puede hacer que nunca se encuentre la soluci√≥n exacta cuando esta existe.\
La principal desventaja de este m√©todo es que es m√°s costoso computacionalmente, ya que se basa en c√°lculos matriciales.
