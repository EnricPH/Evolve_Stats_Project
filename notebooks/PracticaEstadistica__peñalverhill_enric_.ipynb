**`Valores Nulos`**: En caso de que el conjunto de datos contuviese valores nulos, se deber√≠a estudiar si son de tipo:
- Missing at Random (*MAR*): Hay relaci√≥n entre la variable nula respecto a otra/s variable/s.

    - Eg: Sensor de temperatura falla por motivos de isntalaci√≥n, causando valores nulos en valores de temperatura aleatorios.
        
- Missing Not at Random (*MNAR*): Las variables nulas est√°n relacionadas con el valor de la propia variable en s√≠.
    - Eg: Sensor de temperatura falla cuando la temperatura es demasiado alta.

Dependiendo del tipo de valor nulo, de la import√°ncia de la variable y de la cantidad de valores nulos, se pueden aplicar distintas t√©cnicas de imputaci√≥n o simplemente la eliminaci√≥n de las instancias o de la variable completa.



#### 1.2.  **Clasificaci√≥n de Variables:**
En la secci√≥n anterior se ha visto que todas las variables eran de tipo *int*, ahora bien ¬øcu√°les son discretas, cu√°les son binarias y cu√°les podrian pertenecer a variables categoricas ordinales?

Para las variables numericas ordinales, se aprovechar√° el hecho de que ya se sabe, gracias a la descripci√≥n previa del *dataset*, que todas las columnas num√©ricas ordinales no contienen m√°s de 5 "categorias".

````python
def variables_por_tipo(df, tipo):
    """
    Filtra columnas num√©ricas de un DataFrame seg√∫n su tipo:
    'binario', 'ordinal' o 'discreto'.

    Par√°metros
    ----------
    df : pd.DataFrame
        DataFrame de entrada.
    tipo : str
        Tipo de variable a seleccionar. Debe ser uno de:
        {'binario', 'ordinal', 'discreto'}.

    Returns
    -------
    pd.DataFrame
        DataFrame con solo las columnas del tipo solicitado.

    Raises
    ------
    TypeError
        Si df no es un DataFrame de pandas.
    ValueError
        Si tipo no es uno de los valores permitidos.
    """

    # Validaci√≥n de errores
    if not isinstance(df, pd.DataFrame):
        raise TypeError("El primer argumento 'df' debe ser un pandas DataFrame.")

    tipos_validos = {'binario', 'ordinal', 'discreto'}
    if tipo not in tipos_validos:
        raise ValueError(
            f"Tipo '{tipo}' no v√°lido. Debe ser uno de {tipos_validos}."
        )

    # Selecci√≥n solo de columnas num√©ricas
    columnas_seleccionadas = []

    for col in df.columns:
        valores_unicos = len(df[col].unique())

        # BINARIAS
        if tipo == 'binario':
            if valores_unicos == 2:
                columnas_seleccionadas.append(col)

        # ORDINALES (discretas con pocas categor√≠as)
        elif tipo == 'ordinal':
            if valores_unicos > 2 and valores_unicos <= 5:
                columnas_seleccionadas.append(col)

        # DISCRETAS (mayor cardinalidad)
        elif tipo == 'discreto':
            if valores_unicos > 5:
                columnas_seleccionadas.append(col)

    # Output
    print(f"\nVariables de tipo {tipo} seleccionadas:")
    for c in columnas_seleccionadas:
        print('\t¬∑', c)

    return df[columnas_seleccionadas]

# Construios el dataframes para cada uno de los tipos de variables
bin_df = variables_por_tipo(df, 'binario')
display(bin_df.head())

ord_df = variables_por_tipo(df, 'ordinal')
display(ord_df.head())

disc_df = variables_por_tipo(df, 'discreto')
display(disc_df.head())

````

Separar las variables n√∫mericas de esta forma (15 binarias, 3 ordinales, 12 discretas), facilitar√° el estudio para pasos posteriores, ya que tratar de analizar las 30 variables a la vez ser√≠a contraproducente.

- *p-value*: Indica la probabilidad de observar resultados igual de extremos o m√°s a los resultados obtenidos asumiendo que la hip√≥tesis nula (*H0*) es cierta.

#### 2.3.  **Regresi√≥n Lineal (Scikit-Learn):**

**Inferencia vs Predicci√≥n**

Antes de plantear como entrenar el modelo, hay que tener claro el objetivo del estudio.
 - **Inferencia**: El objetivo en la inferencia es entender como las variables seleccionadas afectan a la variable de respuesta mediante la interpretaci√≥n de los coeficientes del modelo, tests de hip√≥tesis, p-values, etc. La inferencia responde a preguntas como:
    - ¬øQue variables estan m√°s relacionadas con la variable de respuesta (variable objetivo)?
    - ¬øCu√°l es la direcci√≥n y magnitud de esta relaci√≥n?
    - ¬øLas relaciones son estad√≠sticamente significativas?
Para escoger un modelo para hacer inferencia, hay que priorizar interpretabilidad antes que precisi√≥n.
 - **Predicci√≥n**: El objetivo de la predicci√≥n es minimizar el error predictivo sobre datos jam√°s vistos por el modelo. La predicci√≥n responde a preguntas como:
    - ¬øQue precisi√≥n obtuvo el modelo?
    - ¬øEl modelo generaliza bien para datos nuevos?
En la predicci√≥n es m√°s com√∫n sacrificar interpretabilidad por precisi√≥n y buena generalizaci√≥n.

*A pesar de que hacer un **estudio predictivo** no es el objetivo de este m√≥dulo de estd√≠stica, a  continuaci√≥n se realizaran ambas t√©cnicas para comprender la distinta metodologia y objetivos. Para obtener una buena comparaci√≥n entre m√©todos, se usaran las misams variables*

**INFERENCIA**

Para esta secci√≥n se **estudiar√° la relaci√≥n** entre las variables `fev1_x10` y `crp_level`, cuya correlaci√≥n √©s del **-0.87**.\
No se usar√° la variable objetivo ya que esta es de car√°cter binario, en este caso se estar√≠a hablando de un problema de **clasificaci√≥n**, no de **regressi√≥n**, y adem√°s no cumpliria la primera, y m√°s importante, suposici√≥n de los modelos de regressi√≥n lineal, la cual √©s linealidad entre variable explicativa y lavariable de respuesta. Forzar un modelo de regresi√≥n lineal en ese contexto producir√≠a resultados enga√±osos y dif√≠cilmente interpretables.

A continuaci√≥n se simular√° un estudio donde `fev1_x10` ser√° la variable de respuesta y `crp_level` ser√° la variable explicativa, on el objetivo de evaluar si existe una **relaci√≥n estad√≠sticamente significativa** entre la inflamaci√≥n y la funci√≥n pulmonar, as√≠ como cuantificar la magnitud y direcci√≥n de dicha relaci√≥n.

Se evaluar√°n m√©tricas y estad√≠sticos propios del an√°lisis inferencial, tales como:

- Coeficientes del modelo (intercepto y pendiente), para interpretar la relaci√≥n entre las variables.
- **Errores est√°ndar**: Cuantifican la variabilidad de los estimadores.
- **p-values**: Contrastar hip√≥tesis sobre la significancia estad√≠stica de los coeficientes.
- **Intervalos de confianza**: Permiten evaluar la robustez de las estimaciones.
- **$R^2$ (R_squared)**: Como medida descriptiva de la proporci√≥n de varianza explicada (no como m√©trica de optimizaci√≥n).

En este caso no ser√° necesario escalar los datos, ya que ambas variables se encuentran en una escala comparable. No obstante, en modelos multivariantes o con variables en escalas distintas, el escalado ser√≠a recomendable mediante t√©cnicas como *MinMaxScaler* o *StandardScaler*.

**PREDICCI√ìN**

Una vez analizada la relaci√≥n entre `fev1_x10` y `crp_level` desde un enfoque **inferencial**, se plantea ahora un estudio con un objetivo distinto: **la predicci√≥n**. En este contexto, el inter√©s principal reside en cuantificar la capacidad del modelo para generalizar y realizar predicciones precisas sobre datos no vistos.

En este estudio, se mantiene el uso de un modelo de regresi√≥n lineal, ya que la variable de respuesta ``fev1_x10`` es discreta y existe una relaci√≥n aproximadamente lineal con ``crp_level``. No se utiliza la variable objetivo binaria del dataset, ya que, como se ha comentado, su modelizaci√≥n corresponder√≠a a un problema de clasificaci√≥n. Para solucionar un problema de clasificaci√≥n usando un modelo lineal, se usan modelos como ***Support Vector Machines (SVM)***

Para evaluar el desempe√±o predictivo del modelo entrenado, se utilizar√°n m√©tricas orientadas al error, que permiten cuantificar la calidad de las predicciones como *RMSE* y *R_squared*, calculada anteriormente.

Para obtener m√°s informaci√≥n a cerca del rendimiento del modelo entrenado, se a√±adiran tambi√©n las siguientes m√©tricas:
- **mean_absolute_error (*MAE*)**: Mide el error absoluto promedio de las predicciones, ismas unidades que la variable de respuesta.
- **mean_absolute_percentage_error (*MAPE*)**: Mide el % de error absoluto promedio de las predicciones.


````python
X = disc_df[['crp_level']]
y = disc_df[['fev1_x10']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Scikit-Learn
lr_model = LinearRegression()
lr_model.fit(X_train,y_train)

# Par√°metros aprendidos
a0 = lr_model.intercept_[0]
a1 = lr_model.coef_[0][0]

# Predicciones
y_pred = lr_model.predict(X_test)

# M√©tricas
rmse = mean_squared_error(y_test, y_pred)**(1/2)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)

print(f'Intercept {a0:.4f}')
print(f'Slope {a1:.4f}')
print(f'RMSE {rmse:.4f}')
print(f'R_Squared {r2:.4f}')
print(f'MAE {mae:.4f}')
print(f'MAPE {mape:.4f}')

plt.figure(figsize=(12,6))
plt.scatter(X_train, y_train, label='Datos de Entrenamiento', alpha=0.7)
plt.scatter(X_test, y_test, label='Datos de Test', color='orange', alpha=0.3)
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regresi√≥n')
plt.legend()
plt.xlabel('crp level (mg/L)')
plt.ylabel('fev1_x10 (litros/segundo)*10')
plt.show()
````

Interpretando los resultados obtenidos de la regresi√≥n:

#### 2.4. **Regresi√≥n logistica:**

A pesar de su nombre, la **Regresi√≥n Logistica** es un modelo de **clasificaci√≥n**, no de **regresi√≥n**, por lo que en este caso si podemos predecir sobre la variable objetivo `lung_cancer_risk`.

Donde la probabilidad de que una instancia de positivo en la variable objetivo, viene dada por:

$$y = \frac{e^{\beta_0 + \beta_1 X}}{1 + e^{\beta_0 + \beta_1 X}}$$

Siendo $\beta_0$ el intercepto y $\beta_1$ la pendiente.

````python
# Variables
X_log = disc_df[['fev1_x10']]
y_log = bin_df['lung_cancer_risk']  # Series, no DataFrame

# Train / Test
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(
    X_log, y_log, test_size=0.3, random_state=1
)

# Modelo con balanceo de clases (ya que la variable objetivo est√° desbalanceada)
log_mdl = LogisticRegression(class_weight='balanced')
log_mdl.fit(X_train_log, y_train_log)

# Predicciones
y_pred_log = log_mdl.predict(X_test_log)
y_prob_log = log_mdl.predict_proba(X_test_log)[:, 1]


# Ordenar para dibujar la curva correctamente
order = np.argsort(X_test_log.values.flatten())
X_sorted = X_test_log.values.flatten()[order]
y_prob_sorted = y_prob_log[order]

# Plot
plt.figure(figsize=(12,6))

plt.scatter(X_log, y_log, label='Datos')
plt.plot(X_sorted, y_prob_sorted, color='red', linewidth=2, label='Probabilidad (Log√≠stica)')
plt.axvline(x=30.35, color = 'green')
plt.xlabel('fev1_x10 (litros/segundo) √ó 10')
plt.ylabel('Probabilidad de Lung Cancer Risk')
plt.legend()
plt.title('Regresi√≥n Log√≠stica con fev1_x10')
plt.show()
````

En este caso no tiene sentido evaluar el modelo usando m√©tricas para modelos de regressi√≥n como lo son: `RMSE` y `R_Squared`, ya que el modelo de **Regresi√≥n Log√≠stica** no predice valores num√©ricos, sino probabilidades de pertenecer a la clase objetivo.

Para los problemas de **clasificaci√≥n** existen otras m√©tricas de evaluaci√≥n de modelos, la mayor√≠a basadas en 4 valores (Teniendo en cuenta que el objetivo es predecir con exito `lung_cancer_risk = 1`):
- **Positivos Reales (TP)**: Individuos de la Clase 1 predecidos como Clase 1
- **Negativos Reales (TN)**: Individuos de la Clase 0 predecidos como Clase 0
- **Positivos Falsos (FP)**: Individuos de la Clase 0 predecidos como Clase 1
- **Neagtivos Falsos (FN)**: Individuos de la Clase 1 predecidos como Clase 0

Estos 4 valores forman la llamada **Matriz de Confusi√≥n**

|   | Real 0 | Real 1 |
|---|---|---|
| **Predicci√≥n 0** | TN  | FN  |
| **Predicci√≥n 1** | FP  | TP  |


Las metricas de evlauaci√≥n m√°s comunes son:

- $Accuracy = \frac{TP+TN}{TP+FP+TN+FN}$: Mide la capacidad de acierto general del modelo

- $Precision = \frac{FP}{TP+FP}$: Mide la capacidad de acierto de la clase objetivo. De todos los clasificacdos como Clase 1, ¬øCu√°ntos realmente lo son?

- $Recall = \frac{FN}{TP+FN}$: Mide la capacidad de detecci√≥n de la clase objetivo. De todos los individuos de Clase 1, ¬øCu√°ntos se puedieron detectar?

- $F1\;Score = \frac{2 Precision Recall}{Precision+Recall}$: Mide el balance entre detecci√≥n (*Recall*) y acierto (*Precision*)

Hay otras m√©tricas como las ***specificity*** o ***ROC-AUC***, pero con las anteriores es m√°s que suficiente en este caso.


Para saber que metricas usar en este caso se debe tener en cuenta que la variable objetivo est√° **desbalanceada** (75% Clase 0 vs 25% Clase 1), por lo que usar la *Accuracy* como referencia no es una buena idea, ya que esta mide todos los aciertos indistintivamente de la clase, por lo que un valor alto de *Accuracy* puede infromar que el modelo capta bien los individuos sanos (la mayor√≠a) pero no los otros.\
En este contexto se necesita un balance entre *Precision* y *Recall*, ja que queremos el menor n√∫mero de **falsos positivos** as√≠ como el menor n√∫mero de **falsos negativos** possible.



````python
precision_fev1 = precision_score(y_test_log, y_pred_log)
recall_fev1 = recall_score(y_test_log, y_pred_log)
f1_score_fev1 = f1_score(y_test_log, y_pred_log)

print(f'Precision: {precision_fev1:.4f}')
print(f'Recall: {recall_fev1:.4f}')
print(f'F1 Score: {f1_score_fev1:.4f}')
````

## PARTE 3: Regresi√≥n Lineal "From Scratch" (Datos Simulados)

**Objetivo:** Demostrar que entiendes la matem√°tica detr√°s de `LinearRegression()` implement√°ndola con NumPy.

Amplia con:

- **Predicci√≥n y evaluaci√≥n**
  Utiliza tu implementaci√≥n para:
  - Realizar predicciones
  - Calcular manualmente el **MSE** y el **R¬≤**, sin usar funciones de Scikit-Learn.

- **Comparaci√≥n con Scikit-Learn**
  Entrena un modelo `LinearRegression()` sobre los mismos datos y compara:
  - Coeficientes
  - M√©tricas de evaluaci√≥n  
  Discute posibles diferencias num√©ricas.


````python
# --- NO TOCAR: GENERACI√ìN DE DATOS ---
np.random.seed(42)
X_mock = 2 * np.random.rand(100, 1)
y_mock = 4 + 3 * X_mock + np.random.randn(100, 1)
# -----------------------------------

def evolve_linear_regression(X, y, iteraciones=16, eta=0.1):
    """
    Calcula la pendiente (w), el intercepto (beta_0),
    el error MSE (mse) y R^2 (r_squared)
    usando las f√≥rmulas de OLS con NumPy.
    Returns: (b, w, b_history, w_history, mse_history, r_squared_history)
    """

    # TU C√ìDIGO AQU√ç
    #Generacion aleatoria de parametros (numpy)
    w = np.random.randn()
    b = np.random.randn()

    n = len(X)

    #Arrays vacios para almacenar los par√°metros y error en cada iteraci√≥n
    w_history = []
    b_history = []
    mse_history = []
    r2_history = []

    #Total Sum of Squares (TSS) constante
    tss = np.sum((y - np.mean(y)) ** 2)

    for i in range(iteraciones):
        #Calcular valor predecido y error (valor real - predecido)
        y_pred = b + w * X
        error = y - y_pred

        #Calculo derivadas de descenso de gradiente
        dw = (-2 / n) * np.sum(X * error)
        db = (-2 / n) * np.sum(error)

        #Ajustar parametros mediante el m√©todo de descenso de gradiente y learning rate (eta)
        w = w - eta * dw
        b = b - eta * db

        # MSE
        mse = np.mean(error ** 2)

        # R_squared 
        rss = np.sum(error ** 2)
        r2 = 1 - rss / tss

        #Guardar c√°lculos
        w_history.append(w)
        b_history.append(b)
        mse_history.append(mse)
        r2_history.append(r2)

    return b, w, b_history, w_history, mse_history, r2_history


# Ejecutamos el modelo
b, w, b_hist, w_hist, mse_history, r2_history = evolve_linear_regression(X_mock, y_mock)


# Modelo Scikit-Learn copn los mismos datos
lr_model = LinearRegression()
lr_model.fit(X_mock, y_mock)

# Par√°metros del modelo de Scikit-Learn
b_sklearn = lr_model.intercept_[0]
w_sklearn = lr_model.coef_[0][0]

# Predicciones del modelo de Scikit-Learn
y_pred_sklearn = lr_model.predict(X_mock)

# M√©tricas del modelo de Scikit -Learn
mse_sklearn = mean_squared_error(y_mock, y_pred_sklearn)
r2_sklearn = r2_score(y_mock, y_pred_sklearn)

# Creaci√≥n de un DataFrame para visualizar mejor la comparaci√≥n entre las dos regresiones
regresiones = pd.DataFrame({'Modelos':['Tu Modelo', 'Modelo Scikit-Learn'], 
        'Intercepto (b)':[b, b_sklearn], 
        'Pendiente estimada':[w, w_sklearn],
        'Error (MSE)': [mse_history[-1], mse_sklearn],
        'Varianza expicada (R_Squared)':[r2_history[-1], r2_sklearn]})

display(regresiones)

#Visualizaci√≥n
plt.figure(figsize=(7, 5))
plt.scatter(X_mock, y_mock, color='blue', label='Datos')

# Rectas intermedias
c = 0
for bi, wi in zip(b_hist[:-1], w_hist[:-1]):
    if c==0:
        plt.plot(X_mock, bi + wi * X_mock, color='gray', alpha=0.3, label='Iteraciones de tu Regresi√≥n')
        c=1
    else:
        plt.plot(X_mock, bi + wi * X_mock, color='gray', alpha=0.3)

# Recta final manual vs LinearRegression() de Scikit-Learn
plt.plot(X_mock, b + w * X_mock, color='red', linewidth=2, label='Tu Regresi√≥n')
plt.plot(X_mock, y_pred_sklearn, color='green', linewidth=2, label='Regresi√≥n scikit-learn')
plt.title('Evoluci√≥n de la Regresi√≥n Lineal')
plt.legend()

plt.tight_layout()
plt.show()

````

Los resultados obtenidos por ambos modelos son muy similares, tanto en los par√°metros estimados como en las m√©tricas de ajuste:

Las peque√±as discrepancias entre valores son esperables y se explican por el **m√©todo de optimizaci√≥n empleado** en cada enfoque.

**``Modelo Manual`` (Descenso de Gradiente)**

- Se basa en un **proceso iterativo**.
- La soluci√≥n final depende de:
  - el n√∫mero de iteraciones,
  - la tasa de aprendizaje, *learning rate* (Œ∑),
  - los valores iniciales aleatorios de los par√°metros.

Utilidad de las medias m√≥viles (moving averages)
Las medias m√≥viles se utilizan para suavizar fluctuaciones de corto plazo y resaltar patrones subyacentes:

**Ventanas cortas (3 meses / 90 d√≠as)**

- Reaccionan r√°pidamente a cambios recientes.
- √ötiles para detecci√≥n temprana de cambios, monitoring operativo o an√°lisis de corto plazo.
- M√°s sensibles al ruido.

**Ventanas intermedias (4‚Äì5 meses / 120‚Äì150 d√≠as)**
- Buen equilibrio entre suavizado y capacidad de respuesta.
- Adecuadas para an√°lisis exploratorio y seguimiento de la tendencia real.

**Ventanas largas (6 meses / 180 d√≠as)**
- Eliminan gran parte del ruido y de la estacionalidad.
- √ötiles para an√°lisis estrat√©gico, planificaci√≥n a largo plazo o comparaci√≥n de ciclos amplios.
- Introducen lag, por lo que reaccionan m√°s tarde a cambios bruscos.

La elecci√≥n del periodo depende del objetivo: detecci√≥n r√°pida vs estabilidad de la se√±al.

Por qu√© descomponer una serie temporal
La descomposici√≥n mediante seasonal_decompose() permite separar la se√±al observada en componentes interpretables, facilitando el an√°lisis, la modelizaci√≥n y el diagn√≥stico de la serie. Es especialmente √∫til para:

- Identificar si la estacionalidad es estable.
- Analizar la tendencia sin interferencia del ruido.
- Verificar si los residuos contienen estructura no explicada.

**Interpretaci√≥n de la descomposici√≥n aditiva**

- Trend
    - Representa el crecimiento estructural de la serie. En este caso se observa una tendencia ascendente suave, consistente con el componente lineal usado en la generaci√≥n de los datos. Confirma que el aumento de las ventas no es puntual ni estacional, sino sistem√°tico.

- Seasonal
    - Muestra un patr√≥n repetitivo y estable en el tiempo, con oscilaciones similares en amplitud y frecuencia. Esto indica una estacionalidad bien definida, coherente con el componente sinusoidal de los datos. Al ser estable, es un buen candidato para ser modelado expl√≠citamente en modelos de series temporales.

- Resid
    - Contiene el ruido no explicado por la tendencia ni la estacionalidad. Se observa que:
    - Est√° centrado alrededor de cero, lo cual es deseable.
    - Presenta cierta estructura ondulada, indicando que a√∫n hay patrones d√©biles no capturados completamente.
    - En un contexto real, residuos con estructura sugerir√≠an mejorar el modelo o ajustar el periodo estacional.

**Relaci√≥n entre medias m√≥viles y descomposici√≥n**

Las medias m√≥viles son una aproximaci√≥n emp√≠rica para aislar la tendencia, mientras que la descomposici√≥n formal separa expl√≠citamente los componentes. En la pr√°ctica:

- Las medias m√≥viles son r√°pidas y √∫tiles para exploraci√≥n.
- La descomposici√≥n es m√°s potente para an√°lisis estructural, forecasting y validaci√≥n de supuestos en modelos temporales.

En conjunto, ambas herramientas permiten entender qu√© parte de la serie es crecimiento real, qu√© parte es c√≠clica y qu√© parte es ruido, lo cual es clave antes de aplicar modelos predictivos como ARIMA, SARIMA o modelos basados en machine learning.
